```python
# Import module
import cryptorocket as cr
```

### Get top100 cmc ranking

```python
# Get a pandas dataframe of current top100 cmc ranking
# And eventually, save it as a CSV file in the folder where your script belong
top100 = cr.api_to_pandas(csv=True)
```

```python
# Check the first lines
top100.head()
```

### Start to compare the top100 ranking between two dates

```python
# Initiate the object with the two datasets from two different dates
old_cmc_data = 'data/cmc_top100_2018-06-10.csv'
fresh_cmc_data = 'data/cmc_top100_2018-09-22.csv'
cmc = cr.CMComparison(old_cmc_data, fresh_cmc_data)
```

```python
# Check which cryptocurrency is new in this top 100
# We expected many because of the period of time between the two datasets
cmc.newcomers_top100()
```

```python
# Similarly, check the list of those which are out
cmc.out_top100()
```

#### Some methods exist to enhanced the freshest CMC ranking

- cmc.delta_ranking()
- cmc.pct_change_market_cap()
- cmc.usd_percent_change()
- cmc.btc_percent_change()

#### One method allows to merge all these information on the freshest CMC ranking dataframe

- cmc.enhanced_fresh_top100()

```python
fresh_top100_enhanced = cmc.enhanced_fresh_top100()
```
